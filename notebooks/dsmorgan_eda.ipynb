## 1. Import Dependencies

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

sns.set_theme(style="whitegrid")
```

## 2. Data Acquisition

```python
df = sns.load_dataset("titanic")
df.head(10)
```

## 3. Initial Data Inspection

```python
print("Shape:", df.shape)
print("\nData types:")
print(df.dtypes)

df.head(10)
```

## 4. Descriptive Statistics

```python
df.describe(include="all").T
```

## 5. Distribution Checks

```python
numeric_cols = df.select_dtypes(include=["number"]).columns
df[numeric_cols].hist(figsize=(12, 8), bins=20)
plt.suptitle("Histograms of Numeric Features")
plt.tight_layout()
plt.show()
```

```python
categorical_cols = df.select_dtypes(include=["object", "category", "bool"]).columns
for col in categorical_cols:
    print(f"\nValue counts for {col}:")
    print(df[col].value_counts(dropna=False))
```

## 6. Data Transformation and Feature Engineering

```python
df = df.rename(columns={"pclass": "passenger_class"})
df["family_size"] = df["sibsp"] + df["parch"] + 1
df["fare_per_person"] = df["fare"] / df["family_size"]

df[["passenger_class", "family_size", "fare_per_person"]].head()
```

## 7. Visual Storytelling

### 7.1 Survival Count by Outcome
- **Goal:** Understand the overall balance of survivors vs. non-survivors.
- **Chart Type:** Count plot

```python
plt.figure(figsize=(7, 4))
sns.countplot(data=df, x="alive", hue="alive", palette="Set2", legend=False)
plt.title("Passenger Survival Counts")
plt.xlabel("Alive")
plt.ylabel("Count")
plt.show()
```

**Story:** The chart shows more passengers did not survive than survived.

### 7.2 Survival Rate by Sex
- **Goal:** Compare survival patterns between male and female passengers.
- **Chart Type:** Bar plot of mean survival

```python
plt.figure(figsize=(7, 4))
sns.barplot(data=df, x="sex", y="survived", hue="sex", palette="pastel", legend=False)
plt.title("Average Survival Rate by Sex")
plt.xlabel("Sex")
plt.ylabel("Survival Rate")
plt.ylim(0, 1)
plt.show()
```

**Story:** Female passengers had a substantially higher survival rate than male passengers, indicating sex was strongly associated with outcomes.

### 7.3 Fare Distribution by Passenger Class
- **Goal:** Evaluate how fare varied across classes and whether class segmentation appears strong.
- **Chart Type:** Box plot

```python
plt.figure(figsize=(8, 4))
sns.boxplot(data=df, x="passenger_class", y="fare", hue="passenger_class", palette="Set3", legend=False)
plt.title("Fare Distribution by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.show()
```

**Story:** First-class passengers generally paid much higher fares, with wider spread and high-end outliers compared with second and third class.

### 7.4 Age Distribution by Survival
- **Goal:** Explore whether age distributions differ between survivors and non-survivors.
- **Chart Type:** Violin plot

```python
age_survival = df.dropna(subset=["age", "alive"])
plt.figure(figsize=(7, 4))
sns.violinplot(data=age_survival, x="alive", y="age", hue="alive", palette="coolwarm", legend=False)
plt.title("Age Distribution by Survival Status")
plt.xlabel("Alive")
plt.ylabel("Age")
plt.show()
```

**Story:** Age distributions overlap considerably, but survivors show a slightly younger central tendency, suggesting age had some influence but was not the only driver.

## 8. Conclusions

This EDA indicates that survival was strongly associated with sex and socioeconomic factors (passenger class and fare). Age appears related to outcomes but with substantial overlap. Additional modeling could quantify feature importance and interactions more formally.
